# Применение метода GridSearch и определение лучшей модели без балансировки признаков.

Так как мы решаем задачу категоризации, то будем применять следующие модели:  
1) Дерево решений    
2) Случайный лес  
3) Логистическая регрессия  
4) Метод k-ближайших соседей

Создадим датафрейм со всеми моделями, которые мы хотим использовать. Далее по циклу из этого датафрейма будем обучать наши модели используя метод кросс валидации.

В методе GridSearch выберем в качестве метрики оценки качетсва модели метрику f1 (scoring = 'f1')

```python
def grid_model_test_output(param_table, slice_fea_train_valid, slice_tar_train_valid, slice_fea_test, slice_tar_test):
    '''
    Возвращает таблицу с с лучшими параметрами моделей, метриками f1, auc_roc. Строит график AUC-ROC для модели 
    с метрикой f1 >= 0.59
    :param_table: исходные данные из которых функция берет модели и параметры моделей, находящиеся в словорях
    :slice_fea_train_valid: выборка, содержащая обучающие и валидационные данные
    :slice_tar_train_valid: выборка, содержащая целевые показатели для обучающей и валидационной выборки
    :slice_fea_test: выборка, содержащая тестовые данные
    :slice_tar_test: выборка, содержащая целевые показатели для тестовой выборки
    ОПИСАНИЕ ПОСЛЕДОВАТЕЛЬНОСТИ РАБОТЫ ФУНКЦИИ: иттерируем param_table по строкам и столбцам. Производим вызов модели через 
    module_imported. Через функцию getattr возвращаем атрибут объекта. В нашем случае атрибут - это модель, объект - это модуль 
    библиотеки. Затем производится подбор лучших параметров модели по метрики f1. Вытаскиваем через best_params_ лучшие 
    параметры модели.Инициализируем модели с лучшими праметрами. Обучаем модели. Делаем предсказания. Считаем метрику f1. 
    Расчитываем вероятность классов. Строим кривую AUC-ROС для модели с f1 >= 0.59. Результат работы функции сохроняется в 
    датафрейм result_df.
    '''
    probabilities_df = []
    f1_df = []
    result_df = []
    for index, row in param_table.iterrows():
        
        module_imported = importlib.import_module(row['module'])
        model_imported = getattr(module_imported, row['model'])
        model = model_imported()
        if verbose:
            print(model)
            
        grid = GridSearchCV(model, row['parameters'], cv = 6, scoring = 'f1')
        grid.fit(slice_fea_train_valid, slice_tar_train_valid)
        if verbose:
            print(grid.best_params_)
    
        try:
            model = model_imported(**grid.best_params_, random_state = 12345)
        except:
            model = model_imported(**grid.best_params_)

        model.fit(slice_fea_train_valid, slice_tar_train_valid)
        predictions = model.predict(slice_fea_test)
           
        f1 = f1_score(slice_tar_test, predictions)
        if verbose:
            print('f1:', f1)
        f1_df.append(f1)
        
        probabilities_test = model.predict_proba(slice_fea_test)
        probabilities_one_test = probabilities_test[:, 1]
        probabilities_df.append(probabilities_one_test)
        auc_roc = roc_auc_score(slice_tar_test, probabilities_one_test)
        if verbose:
            print('auc_roc:', auc_roc)
            print('-----------------------')
        
        result_df.append([f1, auc_roc, row['label'], grid.best_params_])
        
    f1_df = pd.DataFrame(f1_df, columns = ['f1'])
    result_df = pd.DataFrame(result_df, columns = ['Метрика f1', 'AUC-ROC','Название модели', 'Оптимальные параметры']) 
    
    try:
        index_of_f1 = result_df.index[result_df['Метрика f1'] >= 0.59].tolist()
        position = index_of_f1[0]
        fpr, tpr, thresholds = roc_curve(target_test, probabilities_df[position])
        plt.figure()
        plt.plot(fpr, tpr)
        plt.plot([0, 1], [0, 1], linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.0])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC-кривая')
        plt.show()  
        return result_df.sort_values(by = 'Метрика f1', ascending = False)
        
    except:
        return result_df.sort_values(by = 'Метрика f1', ascending = False)
```

## Балансировка признаков по методу downsampling вручную
<a id="balance_downsampling"></a>

Посчитаем частотность отрицательных целевых признаков

```python
target = df['Exited']
#Признаки объектов с целевым признаком равном нулю
features_train_and_valid_zeroes = features_train_and_valid[target == 0]
#Целевой признак объектов, у которых ответы равны нулю
target_train_and_valid_zeroes = target_train_and_valid[target == 0]

#Признаки объектов с целевым признаком равным единице
features_train_and_valid_ones = features_train_and_valid[target == 1]
#Целевой признак объектов, у которых ответы равны единице
target_train_and_valid_ones = target_train_and_valid[target == 1]

print(f"Кол-во объектов с целевым признаком равным нулю: {len(features_train_and_valid_zeroes)}")
print(f"Кол-во объектов с целевым признаком равным единице: {len(features_train_and_valid_ones)}")
print(f"Целевой признак объектов, у которых ответы равны нулю: {len(target_train_and_valid_zeroes)}")
print(f"Целевой признак объектов, у которых ответы равны единице: {len(target_train_and_valid_ones)}")
print(f"Отношение отрицательных признаков к положительным: {(5605 / 1395)}")
```

Таким образом, нам необходимо уменьшить кол-во отрицательных значений в 4 раза. frac коэффциент равен 0.25

```python
#Применим функцию sample для случайного выброса значений признаков для тренировочной и валидац выборок
features_train_and_valid_downsampled = pd.concat(
        [features_train_and_valid_zeroes.sample(frac = 0.25)] + [features_train_and_valid_ones])

#Применим функцию sample для случайного выброса значений целевого признака через для тренировочной и валидац выборок
target_train_and_valid_downsampled = pd.concat(
        [target_train_and_valid_zeroes.sample(frac = 0.25)] + [target_train_and_valid_ones])


#Применим функцию shuffle для перемешивания значений в наших выборках.
features_train_and_valid_downsampled, target_train_and_valid_downsampled = shuffle(
        features_train_and_valid_downsampled, target_train_and_valid_downsampled)
```

## Балансировка признаков по методу upsampling в ручную
<a id="balance_upsampling"></a>

Таким образом, нам необходимо увеличть кол-во положительных значений в 2 раза.

```python
#Применим функцию для увеличения количетсва значений признаков для тренировочной и валидац выборок
features_train_and_valid_upsampled = pd.concat([features_train_and_valid_zeroes] + [features_train_and_valid_ones] * 2)

#Применим функцию для увеличения количетсва значений целевых признаков для тренировочной и валидац выборок
target_train_and_valid_upsampled = pd.concat([target_train_and_valid_zeroes] + [target_train_and_valid_ones] * 2)
    
#Применим функцию shuffle для перемешивания значений в наших выборках.    
features_train_and_valid_upsampled, target_train_and_valid_upsampled = shuffle(
       features_train_and_valid_upsampled, target_train_and_valid_upsampled)
```
