# Предобработка данных.

### Округление минут разговоров и трафика
<a id="round"></a>

В датафрейме **data.calls** окргулим данные в столбце duration, которые принимают значения в интервале (0.0;1.0) до 1.0 минуты. В датафрейме **data.internet** окргулим данные в столбце mb_used, которые принимают знаения в интервале (0.0;1.0) до 1.0 Мб.

```python
data_calls['duration'] = np.ceil(data_calls['duration'])
```

```python
data_internet['mb_used'] = np.ceil(data_internet['mb_used'])
```

### Добавление столбца месяц в таблицы data_calls, data_internet, data_messages
<a id="round"></a>

Добавим в каждую из таблиц столбец месяц. В дальнейщем будем группирровать данные и сводить в одну таблицу через столбцы user_id, месяц.

```python
data_calls['месяц'] = pd.DatetimeIndex(data_calls['call_date']).month 
display(data_calls)
```

```python
data_internet['месяц'] = pd.DatetimeIndex(data_internet['session_date']).month 
data_messages['месяц'] = pd.DatetimeIndex(data_messages['message_date']).month 
```

### Сведение данных в одну таблицу
<a id="round"></a>

```python
user_calls = data_calls.groupby(['user_id', 'месяц']).agg({'duration':'sum', 'id':'count'}).reset_index()
user_messages = data_messages.groupby(['user_id', 'месяц']).agg({'id':'count'}).reset_index()
user_internet = data_internet.groupby(['user_id', 'месяц']).agg({'mb_used':'sum'}).reset_index()
```

Посмотрим на результаты группировки

```python
display(user_calls)
```

Получили, что звонки с длительностью 0.0 минут были проссумировны с остальными звонками (это учитывать в дальнейщем - когда будем объединять данные). То есть, другими словами, мы их не увидим в общей статистики (что положительно для нас). С точки зрения бизнесса в этом нет чего страшного, так как это несовершенные по разным причинам звонки и минуты по данным звонкам не засчитывались в использование, что не как не влияло на выручку компании. 

```python
display(user_messages)
```

```python
display(user_internet)
```

```python
data_total['мин_больше_тариф'] = data_total['продолжительность_звонка'] - data_total['minutes_included']
data_total['смс_больше_тариф'] = data_total['количество_смс'] - data_total['messages_included']
data_total['Гб_больше_тариф'] = data_total['Гб_использовано'] - data_total['gb_per_month_included']
```

Далее напишем функции, которые будут проверять значения в столбцах мин_больше_тариф, смс_больше_тариф, Гб_больше_тариф. Если значения меньше нуля, то клиент тарифного плана уложился в тарифный план - вернем значение ноль; если значение больше нуля, то клиент начал пользоватсья услугами сверх тарифного плана.

```python
def minutes_usage(data_total):
    """Функуция для определения объема переиспользования минут по тарифам."""
    if data_total['мин_больше_тариф'] <= 0:
        return 0
    return data_total['мин_больше_тариф']

data_total['мин_больше_тариф'] = data_total.apply(minutes_usage, axis = 1)
```

```python
def sms_usage(data_total):
    """Функуция для определения объема переиспользования смс по тарифам."""
    if data_total['смс_больше_тариф'] <= 0:
        return 0
    return data_total['смс_больше_тариф']

data_total['смс_больше_тариф'] = data_total.apply(sms_usage, axis = 1)
```

```python
def gb_usage(data_total):
    """Функуция для определения интернет трафика по тарифам."""
    if data_total['Гб_больше_тариф'] <= 0:
        return 0
    return data_total['Гб_больше_тариф']

data_total['Гб_больше_тариф'] = data_total.apply(gb_usage, axis = 1)
```

### Удаление выбросов в данных по диаграммам размаха
<a id="outliners"></a>

Удалим выбросы у столбцов 'продолжительность_звонка', 'количествово_звонков', 'количество_смс', 'Мб_использовано'. Введем функцию iqr для определения статистических параметров выборки и интервалов нахождения выбросов по диаграмме размаха.

```python
def iqr(df, column):
    """Функция для определения выбрососв по методу диаграммы размаха.
    
    Ключевые аргументы:
        df - датафрейм, данные которого мы будем использовать
        column - по данному столбцу мы будем считать границы выбросов
        
    """
    print(df[column].describe())
    iqr = df[column].describe()[6] - df[column].describe()[4]
    low_bond = df[column].describe()[4] - 1.5 * iqr # считаем нижнюю границу
    if low_bond < 0: # бывает так, что нижняя граница уходит в минус, значит ей можно присвоить минимум признака
        low_bond = df[column].describe()[3]
    upper_bond = df[column].describe()[6] + 1.5 * iqr # верхняя гранциа
    print('Нижняя граница:', low_bond)
    print('Верхняя граница:', upper_bond) 
```

```python
for column in data_total.columns.tolist()[2:6]:
    iqr(data_total, column)
```

### Проверка гипотез
<a id="hypothesis_check_it_out"></a>

#### Проверка гипотез средней выручки пользователей тарифов "Ультра" и "Смарт"
<a id="rev_for_ultra_smart"></a>

Выдвенем нулевую гипотезу, что среднее значение выручки на одного клиента в месяц по тарифу "Ультра" = средней выручке на одного клиента в месяц по тарифу "Смарт".   
Среднее значение для data_total_ultra['выручка_на_одного_клиента_рублей_ultra'] = 1950 рублей/клиент месяц 
Среднее значение для data_total_smart['выручка_на_одного_клиента_рублей_smart'] = 1038 рублей/клиент месяц  
Дургими словами мы проверим гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

```python
data_total_ultra['выручка_на_одного_клиента_рублей_ultra'].describe()
```

```python
data_total_smart['выручка_на_одного_клиента_рублей_smart'].describe()
```

Применим метод scipy.stats.ttest_ind. В качестве аргумента equal_var выберем False, так как средниквадратичные отклонения очень сильно отличаются: 100.1 рубль и 772.0 рубля. Критерием выбора аргумента equal_var будет являться разность дисперсий. Дисперсия в свою очередь равна квадрату стандартного отклонения. Таким образом применяем аргумент equal_var = False

```python
alpha = 0.05  # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(data_total_ultra['выручка_на_одного_клиента_рублей_ultra'], 
                      data_total_smart['выручка_на_одного_клиента_рублей_smart'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else: 
    print("Принимаем нулевую гипотезу")
```

Полученное значение говорит нам о том, что p-value очень сильно мало по сравнению с порогом значимости alpha = 0.05. Необходимо отбросить нулевую гипотезу. То есть: **средняя выручка на одного клиента в месяц по тарифу "Смарт" не равна средней выручке на одного клиента в месяц по тарифу "Ультра"**

#### Проверка гипотез средней выручки пользователей из Москвы и дургих регионов
<a id="rev_for_ultra_smart"></a>

Выдвенем нулевую гипотезу, что среднее значение выручки на одного клиента в месяц по региону Москва = средней выручке на одного клиента в месяц по остальным регионам.   
Среднее значение для data_total_moscow['выручка_на_одного_клиента_рублей_москва'] = 1750 рублей/клиент месяц 
Среднее значение для data_total_wo_moscow['выручка_на_одного_клиента_рублей_регионы'] = 1587 рублей/клиент месяц  
Дургими словами мы проверим гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

```python
data_total_moscow['выручка_на_одного_клиента_рублей_москва'].describe()
```

```python
data_total_wo_moscow['выручка_на_одного_клиента_рублей_регионы'].describe()
```

Применим метод scipy.stats.ttest_ind. В качестве аргумента equal_var выберем False, так как средниквадратичные отклонения отличаются на 6.66%: 694.7 рубля и 741.0 рубль. Критерием выбора аргумента equal_var будет являться разность дисперсий. Дисперсия в свою очередь равна квадрату стандартного отклонения и разница по дисперсиям будет дажде более 6.66%. Таким образом применяем аругмент equal_var = False.

```python
alpha = 0.05  # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(data_total_moscow['выручка_на_одного_клиента_рублей_москва'], 
                      data_total_wo_moscow['выручка_на_одного_клиента_рублей_регионы'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else: 
    print("Принимаем нулевую гипотезу")
```

Полученное значение говорит нам о том, что p-value больше alpha = 0.05. Необходимо принять нулевую гипотезу. То есть: **среднее значение выручки на одного клиента в месяц по региону Москва = средней выручке на одного клиента в месяц по остальным регионам.**
